# Sampling

```python
from imblearn.over_sampling import SMOTE
```

```python
sample=SMOTE(random_state=42)
```

```python
X = df.drop('Diagnosis',axis=1)
y=df['Diagnosis']
```

```python
X_resampled,y_resampled=sample.fit_resample(X,y)
df_sampled=pd.DataFrame(X_resampled,columns=df.columns[:-1])
df_sampled['Diagnosis']=y_resampled
```

```python
df_sampled
```

# Feature Selection

```python
target_corr=corr['Diagnosis'].abs().sort_values(ascending=False)
target_corr
```

```python
top_5_features=target_corr.index[1:6].tolist()
top_5_features
```

```python
from sklearn.pipeline import Pipeline
from sklearn import set_config
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestClassifier
```

```python
pipe = make_pipeline(RandomForestClassifier())
pipe
```

# Splitting the dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X_giving_for_splitting=X_resampled[top_5_features]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_giving_for_splitting,y_resampled,test_size=0.2,random_state=42,stratify=y_resampled)
```

```python
print(f"shape of X_train: {X_train.shape}")
print(f"shape of X_test: {X_test.shape}")
print(f"shape of y_train: {y_train.shape}")
print(f"shape of y_test: {y_test.shape}")
```

```python
pipe.fit(X_train,y_train)
pipe
```

```python
from sklearn.metrics import accuracy_score, classification_report , confusion_matrix
```

```python
y_pred=pipe.predict(X_test)
```

```python
y_pred.shape
```

```python
accuracy=accuracy_score(y_test,y_pred)
print(f"accuracy score = {accuracy}")
```

```python
classification=classification_report(y_test,y_pred)
print(f"classification report = {classification}")
```

```python
confusion=confusion_matrix(y_test,y_pred)
print(f"confusion marix = {confusion}")
```
